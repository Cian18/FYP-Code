Since dataset is quite small I will create a custom linear regression function.

After analysing the trajectories, some of the angles wrap around from $-180^\circ$ to $180^\circ$. This effect will treat some of the points as outliers even if they are not.

A solution to this is to transform $\theta$ into x and y, then perform regression on x and y separately.
- x = $cos(\theta)$ = $A_0+A_1t_{norm}$
- y = $sin(\theta)$ = $B_0+B_1t_{norm}$

then after performing the regression we can transform back into theta with the arctan2 function.

To summarise, the regression equation will be:
- $\theta=arctan2(\frac{B_0+B_1t_{norm}}{A_0+A_1t_{norm}})$

```python
#This function will take in normalized time and the joint angle.
#The ouptut is the coefficients ((A0, A1),(B0, B1)) for a liear function of the form Y = A0 + A1X
def L_Regression_ang(t_norm,theta):
    #first transform theta into x and y.
    x = np.cos(theta)
    y = np.sin(theta)
    
    X = np.stack((x,y)).T
    
    #solve a matrix equation to get the A0 and A1 coefficients
    M_x = np.array([[len(t_norm),np.sum(t_norm)],[np.sum(t_norm),np.sum(t_norm**2)]])
    V_x = np.array([np.sum(X[:,0]),np.sum(X[:,0]*t_norm)])
    C_x = np.linalg.solve(M_x,V_x)
    
    #solve a matrix equation to get the B0 and B1 coefficients
    M_y = np.array([[len(t_norm),np.sum(t_norm)],[np.sum(t_norm),np.sum(t_norm**2)]])
    V_y = np.array([np.sum(X[:,1]),np.sum(X[:,1]*t_norm)])
    C_y = np.linalg.solve(M_y,V_y)
    
    #combine the coefficients
    coeffs = np.stack((C_x,C_y))
    
    return coeffs
```

Also create a linear regression function for the non angular data.

```python
def L_regression(t_norm, X):
    M = np.array([[len(t_norm),np.sum(t_norm)],[np.sum(t_norm),np.sum(t_norm**2)]])
    v = np.array([np.sum(X),np.sum(X*t_norm)])
    coeffs = np.linalg.solve(M,v)
    return coeffs
```

After analysing all the plots, I have concluded that the eccentric contraction of the vertical hip to knees trajectory should be fitted with a quadratic regression function.

Therefore, create a quadratic regression function.

```python
#This function will take in normalized time and the joint angle.
#The ouptut is the coefficients ((A0, A1),(B0, B1)) for a linear function of the form Y = A0 + A1X
def Q_Regression_ang(t_norm,theta):
    #first transform theta into x and y
    x = np.cos(theta)
    y = np.sin(theta)
    
    #solve a matrix equation to get the A0 and A1 coefficients
    M_x = np.array([[len(t_norm),np.sum(t_norm),np.sum(t_norm**2)],[np.sum(t_norm),np.sum(t_norm**2),np.sum(t_norm**3)],[np.sum(t_norm**2),np.sum(t_norm**3),np.sum(t_norm**4)]])
    V_x = np.array([np.sum(x),np.sum(x*t_norm),np.sum(x*t_norm**2)])
    C_x = np.linalg.solve(M_x,V_x)
    
    #solve a matrix equation to get the B0 and B1 coefficients
    M_y = np.array([[len(t_norm),np.sum(t_norm),np.sum(t_norm**2)],[np.sum(t_norm),np.sum(t_norm**2),np.sum(t_norm**3)],[np.sum(t_norm**2),np.sum(t_norm**3),np.sum(t_norm**4)]])
    V_y = np.array([np.sum(y),np.sum(y*t_norm),np.sum(y*t_norm**2)])
    C_y = np.linalg.solve(M_y,V_y)
    
    #combine the coefficients
    coeffs = np.stack((C_x,C_y))
    
    return coeffs
```

```python
def Q_regression(t_norm, X):
    M = np.array([[len(t_norm), np.sum(t_norm), np.sum(t_norm**2)],[np.sum(t_norm), np.sum(t_norm**2), np.sum(t_norm**3)],[np.sum(t_norm**2), np.sum(t_norm**3), np.sum(t_norm**4)]])
    v = np.array([np.sum(X), np.sum(t_norm*X), np.sum(t_norm**2*X)])
    coeff = np.linalg.solve(M,v)
    return coeff
```

## Prepare for the Fuzzy Logic
In preparation for the feedback section, I will calculate the signed error of each datapoint from the corresponding one on the regression model. Then I will calculate the mean and standard deviation of this error.

Note that there is a complication with finding the error for some of the angles, if the angle wraps around from close to -180 to close to 180 then this will cause a massive error, therefore I will calculate the error in x and the error in y and then use arctan2 to find the corresponding angular error.

```python
#Initiate an empty list that has a nested list for each feature
err_con = [[],[],[],[]]

#First loop through the concentric contraction.
for i in range(len(con)):
    #define the normalised time scale
    t = np.linspace(0,1,len(con[i]))
    keys = list(coeffs_con.keys())

    for j in range(con[i].shape[1]):
        #for angles
        if j==0:
            t_lin = np.stack((np.ones(len(t)),t))
            #find the value of angle predicted by regression model (it is in radians)
            corr = np.arctan2(coeffs_con[keys[j]][1]@t_lin, coeffs_con[keys[j]][0]@t_lin)
            #find the error
            err = con[i][:,j]-corr
            #deal with the wrap-around problem by finding the minimum angular error
            err = (err+np.pi)%(2*np.pi)-np.pi
            
        elif j==1:
            t_lin = np.stack((np.ones(len(t)),t,t**2))
            corr = coeffs_con[keys[j]]@t_lin
            #calculate the error
            err = con[i][:,j]-corr
        #for non-angles
        elif j==2:
            t_lin = np.stack((np.ones(len(t)),t))
            corr = coeffs_con[keys[j]]@t_lin
            #calculate the error
            err = con[i][:,j]-corr
        else:
            t_quad = np.stack((np.ones(len(t)),t,t**2))
            corr = coeffs_con[keys[j]]@t_quad
            #calculate the error
            err = con[i][:,j]-corr
        
        #loop through each point error that was just calculated and append it to the corresponding nested list
        for k in range(len(err)):
            err_con[j].append(err[k])
        
#turn the error into an array
err_con = np.array(err_con)
```

```python
err_con_mean = np.round(np.mean(err_con, axis=1),3)
err_con_std = np.round(np.std(err_con, axis=1),3)
np.round(err_con_mean*180/np.pi,2),np.round(err_con_std*180/np.pi,3)
```
